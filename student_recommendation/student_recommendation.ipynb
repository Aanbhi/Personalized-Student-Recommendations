```python
# ✅ Step 2: Fetch Quiz Data from API
import requests
import json

QUIZ_API_URL = "https://api.jsonserve.com/rJvd7g"

def fetch_quiz_data():
    response = requests.get(QUIZ_API_URL)
    if response.status_code == 200:
        return response.json()
    else:
        print("Failed to fetch quiz data:", response.status_code)
        return None

# Fetch and inspect data
quiz_data = fetch_quiz_data()
print(json.dumps(quiz_data, indent=2))  # Print data for quick inspection
```

```python
import pandas as pd

def process_data(data):
    print("Data Keys:", data.keys())  # Debug: Show available keys

    # Extract quiz details if it's nested inside 'quiz'
    if 'quiz' in data and isinstance(data['quiz'], dict):
        quiz_details = data['quiz']
    else:
        quiz_details = data  # If 'quiz' is missing, use data as-is

    # Convert to DataFrame (Wrap in a list if it's a single dictionary)
    df = pd.DataFrame([quiz_details])

    # Debug: Print available columns
    print("Columns in DataFrame:", df.columns)

    # Handle missing 'user_id' safely
    if 'user_id' not in df.columns:
        print("Warning: 'user_id' not found in data.")
        df['user_id'] = None  # Add a placeholder column

    # Ensure necessary columns exist before calculating accuracy
    if 'correct_answers' in df.columns and 'total_questions' in df.columns:
        df['accuracy'] = (df['correct_answers'] / df['total_questions']) * 100
    else:
        print("Warning: Missing required columns for accuracy calculation.")

    return df

# Process Data
if quiz_data:
    df = process_data(quiz_data)
    print(df.head())  # Display processed data
```
